```python
# Importacin de las libreras necesarias
import numpy as np  # Biblioteca para operaciones numricas y manejo de arreglos
import matplotlib.pyplot as plt  # Biblioteca para la generacin de grficos
from ipywidgets import interact  # Herramienta para crear interfaces interactivas en Jupyter Notebook

# Definicin de una funcin interactiva que permite modificar las frecuencias de las seales de entrada
@interact(f1=(0, 50, 1), f2=(0, 50, 1), f3=(0, 50, 1))  # Se definen los rangos y pasos de las frecuencias
def sin(f1=10, f2=6, f3=4):
    """
    Funcin que genera y grafica dos seales senoidales y su convolucin.

    Parmetros:
    f1 (int): Frecuencia de la primera componente de la seal compuesta.
    f2 (int): Frecuencia de la segunda componente de la seal compuesta.
    f3 (int): Frecuencia de la segunda seal individual.
    """

    # Definir el vector de tiempo para la simulacin
    t = np.linspace(0, 1, 1000)  # Se genera un arreglo de 1000 puntos en el intervalo [0,1] segundos

    # Generacin de las seales de entrada
    s1 = np.sin(2 * np.pi * t * f1) + np.sin(2 * np.pi * t * f2)  # Seal compuesta por dos senoidales
    s2 = np.sin(2 * np.pi * f3 * t)  # Seal senoidal simple

    # Clculo de la convolucin de las seales s1 y s2
    conv = np.convolve(s1, s2)  # Se obtiene la respuesta de convolucin
    tc = np.linspace(0, 1, len(conv))  # Ajuste del vector de tiempo para la convolucin

    # Creacin de la figura y subgrficos
    fig = plt.figure(figsize=(10, 6))  # Definir el tamao de la figura

    # Configuracin de los subgrficos en una cuadrcula 2x2
    ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=1)  # Subgrfico para la seal 1
    ax2 = plt.subplot2grid((2, 2), (0, 1), colspan=1)  # Subgrfico para la seal 2
    ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)  # Subgrfico para la convolucin

    # Graficar la seal 1
    ax1.plot(t, s1, color="blue")
    ax1.set_title("Seal 1")
    ax1.set_xlabel("Tiempo (s)")
    ax1.set_ylabel("Amplitud")
    ax1.grid(True)  # Habilitar la cuadrcula para mejor visualizacin

    # Graficar la seal 2
    ax2.plot(t, s2, color="blue")
    ax2.set_title('Seal 2')
    ax2.set_xlabel("Tiempo (s)")
    ax2.set_ylabel("Amplitud")
    ax2.grid(True)

    # Graficar la convolucin
    ax3.plot(tc, conv, color="black")
    ax3.set_title('Convolucin')
    ax3.set_xlabel("Tiempo (s)")
    ax3.set_ylabel("Amplitud")
    ax3.grid(True)

    # Ajuste del diseo de los grficos para evitar superposicin
    plt.tight_layout()

    # Mostrar la figura con las grficas
    plt.show()
```

```python
# Importacin de libreras necesarias
import numpy as np  # Para el manejo de arreglos numricos
import matplotlib.pyplot as plt  # Para la generacin de grficos
from ipywidgets import interact  # Para la creacin de widgets interactivos

# Definicin de una funcin interactiva que permite modificar parmetros de la seal
@interact(f1=(0, 120, 1), f2=(0, 120, 1), f3=(0, 120, 1), t_c=(0, 1, 0.001), width=(0.01, 1, 0.001))
def sig_cut(f1=10, f2=20, f3=60, t_c=0.5, width=0.2):
    """
    Genera y grafica seales senoidales con parmetros ajustables mediante widgets interactivos.
    Permite la modificacin de:
      - f1, f2: Frecuencias de la seal compuesta.
      - f3: Frecuencia de la seal recortada en un intervalo de tiempo.
      - t_c: Centro del intervalo donde la seal f3 est activa.
      - width: Duracin del intervalo donde la seal f3 est activa.

    Se grafican tres seales:
      1. Seal compuesta por dos senoidales (f1 y f2).
      2. Seal f3 activa solo en un intervalo especfico.
      3. Convolucin entre ambas seales.
    """

    # Definir el vector de tiempo en el intervalo [0, 1] con 1000 muestras
    t = np.linspace(0, 1, 1000)

    # Inicializar la seal con ceros
    sig = np.zeros(len(t))

    # Definir los lmites del intervalo [t1, t2] donde la seal f3 estar activa
    t1 = max(0, t_c - width / 2)  # Evita valores negativos para t1
    t2 = min(1, t_c + width / 2)  # Evita que t2 supere 1

    # Generacin de la seal compuesta por dos senoidales con frecuencias f1 y f2
    sen_compuesta = np.sin(2 * np.pi * t * f1) + np.sin(2 * np.pi * t * f2)

    # Aplicar la mscara booleana para activar la seal f3 solo en el intervalo [t1, t2]
    mask = (t >= t1) & (t <= t2)
    sig[mask] = np.sin(2 * np.pi * t[mask] * f3)  # Seal activa en el intervalo definido

    # Clculo de la convolucin de la seal compuesta con la seal recortada
    convolucion = np.convolve(sen_compuesta, sig, mode='same')  # Se obtiene la respuesta de convolucin
    tconv = np.linspace(0, 1, len(convolucion))  # Ajuste del vector de tiempo para la convolucin

    # Creacin de la figura y subgrficos
    fig = plt.figure(figsize=(10, 6))  # Definir el tamao de la figura

    # Configuracin de los subgrficos en una cuadrcula 2x2
    ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=1)  # Subgrfico para la seal 1
    ax2 = plt.subplot2grid((2, 2), (0, 1), colspan=1)  # Subgrfico para la seal 2
    ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)  # Subgrfico para la convolucin

    # Graficar la seal 1 (compuesta por f1 y f2)
    ax1.plot(t, sen_compuesta, color="blue")
    ax1.set_title("Seal compuesta (f1 + f2)")
    ax1.set_xlabel("Tiempo (s)")
    ax1.set_ylabel("Amplitud")
    ax1.grid(True)  # Habilitar la cuadrcula para mejor visualizacin

    # Graficar la seal 2 (seal f3 activa en un intervalo)
    ax2.plot(t, sig, color="blue")
    ax2.set_title('Seal recortada (f3 en intervalo)')
    ax2.set_xlabel("Tiempo (s)")
    ax2.set_ylabel("Amplitud")
    ax2.grid(True)

    # Graficar la convolucin entre ambas seales
    ax3.plot(tconv, convolucion, color="black")
    ax3.set_title('Convolucin de las seales')
    ax3.set_xlabel("Tiempo (s)")
    ax3.set_ylabel("Amplitud")
    ax3.grid(True)

    # Ajuste del diseo de los grficos para evitar superposicin
    plt.tight_layout()

    # Mostrar la figura con las grficas
    plt.show()
```
